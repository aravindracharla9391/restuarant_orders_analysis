### Load restaurants data (SQL)

```python

import sqlite3
import pandas as pd

db_path = r"C:\Users\saikr\Downloads\restaurants.db"
# connect with mysql 
conn = sqlite3.connect(db_path)
cursor = conn.cursor()
```

```python
sql_path = r"C:\Users\saikr\Downloads\intern_restuarant.sql"

with open(sql_path, "r", encoding="utf-8") as file:
    sql_script = file.read()
```

```python
clean_sql = []

for line in sql_script.splitlines():
    l = line.strip().lower()
    if l.startswith("create database"):
        continue
    if l.startswith("use "):
        continue
    if l.startswith("drop database"):
        continue
    clean_sql.append(line)

clean_sql_script = "\n".join(clean_sql)
```

```python
cursor.executescript(clean_sql_script)
conn.commit()
```

```python
pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';",conn)
```

```python
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants
```

```python
restaurants.head() # to get first 5 rows from restuarnts dataset
```

```python
# checking the no of rows and coumns and to get first 5 rows of datset
print(restaurants.shape)
print(restaurants.head())
```

```python
# we are merging two dataset users and orders by using merge 
orders_users = pd.merge(orders,users,on="user_id",how="left")
```

```python
# from above we merged orders and users into one variable called orders_users now we can merge with this resturants inorder to get complete 3 tables together
final_df = pd.merge(orders_users,restaurants,on="restaurant_id",how="left")
```

```python
# Converting order_date column into datetime format for monthly analysis
final_df['order_date'] = pd.to_datetime(final_df['order_date'])
# Grouping data by month and counting total orders
final_df.groupby(final_df['order_date'].dt.month)['order_id'].count()
```

```python
# Which city has the highest total revenue from Gold members?
(final_df[final_df['membership'] == 'Gold'].groupby('city')['total_amount'].sum().sort_values(ascending=False))
```

```python
#Which cuisine has the highest average order value across all orders?

(final_df.groupby('cuisine')['total_amount'].mean().sort_values(ascending=False))
```

```python
#How many distinct users placed orders worth more than ₹1000 in total (sum of all their orders)?
user_spend = (final_df.groupby('user_id')['total_amount'].sum())
user_spend[user_spend > 1000].count()
```

```python
#Which restaurant rating range generated the highest total revenue?

bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0–3.5', '3.6–4.0', '4.1–4.5', '4.6–5.0']

final_df['rating_range'] = pd.cut(final_df['rating'],bins=bins,labels=labels,include_lowest=True)
```

```python
(
    final_df.groupby('rating_range')['total_amount']
    .sum()
    .sort_values(ascending=False)
)
```

```python
#Among Gold members, which city has the highest average order value?

(final_df[final_df['membership'] == 'Gold'].groupby('city')['total_amount'].mean().sort_values(ascending=False))
```

```python
#Which cuisine has the lowest number of distinct restaurants but still contributes significant revenue?
restaurant_count = (final_df.groupby('cuisine')['restaurant_id'].nunique())
cuisine_revenue = (final_df.groupby('cuisine')['total_amount'].sum())
summary = pd.concat( [restaurant_count, cuisine_revenue], axis=1)
summary.columns = ['restaurant_count', 'total_revenue']
summary.sort_values(['restaurant_count', 'total_revenue'],ascending=[True, False])
```
